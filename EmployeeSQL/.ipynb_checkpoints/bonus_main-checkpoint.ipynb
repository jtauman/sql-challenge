```python
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
#from password import password
```

```python
#database_uri = 'postgres+psycopg2://postgres:password@localhost:5432/Employee_db'
```

```python
engine = create_engine('postgresql://postgres:1Pgx2hot!@localhost/Employee_db')
```

```python
conn = engine.connect()
data = pd.read_sql("SELECT * FROM Employee",conn)
```

```python
data.head()
```

```python
salary_data = pd.read_sql("SELECT * FROM Salaries",conn)
salary_data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
salary = salary_data["salary"]
salary.hist(bins = 50)
plt.title("Histogram of Employee Salaries")
plt.xlabel("Salary ($)")
plt.ylabel("Frequency")
plt.savefig("images/histogram.png")
plt.show()
```

```python
#create dataframe of average salary by title
merged_data = pd.merge(data, salary_data, on = 'emp_no')
merged_data.head(20)
```

```python
titles_data = pd.read_sql("SELECT * FROM Titles",conn)
titles_data.head()
```

```python
merged_data_final = pd.merge(merged_data, titles_data, left_on = 'emp_title_id', right_on = 'title_id')
merged_data_final.head()
```

```python
title_group = merged_data_final.groupby(["title"])
avg_salary = round(title_group["salary"].mean(),2)
avg_salary_df = pd.DataFrame({"Average Salary": avg_salary})
avg_salary_df
```

```python
avg_salary_df.plot(kind = "bar")
plt.title("Average Salary")
plt.ylabel("Salary ($)")
plt.savefig("images/avg_salary.png")
plt.show()
```
